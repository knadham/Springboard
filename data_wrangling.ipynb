# Understanding the sharing behind Airbnb's sharing economy


Sharing economy platforms have become increasingly popular in the past few years, and they are dramatically changing traditional businesses and daily life. One of the most prominent examples of the sharing economy is Airbnb, an online marketplace for short-term accommodations. Found in 2008, Airbnb brings together people with room to spare and people looking for a place to stay. However, in it's almost decade of existence, many feel that very little is left of the platform's original concept of “sharing” unused spaces with visitors and establishing a connection remains. This is evidenced by the high proportion of entire home listings on the site as well as the growing presence of "professional hosts"-- users with more than one listing on the site. In response to this criticism, Airbnb has published various reports on how the service is used and the impact it has users and communities. This exercise focuses on the presence of Airbnb in different cities across Europe and understanding how the platform is being used. 

The following questions were considered

-How much how much “sharing,” defined as private and shared room listings, takes place on the sharing economy platform?
<br>
-Are Airbnb hosts mainly “regular people” looking for supplemental income or culture exchange or rather professionals?
<br>
-Are most Airbnb propoerties outside main hotel districts?
